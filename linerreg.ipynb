```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
train_csv='train_data.csv'
val_csv='val_data.csv'
```

```python
df=pd.read_csv(train_csv)
print(df.head())
```

```python
W=np.random.rand()
B=np.random.rand()

print(W,B)
```

```python
x_train=df["X"].to_list()
y_train=df["Y"].to_list()

print(x_train,'\n',y_train)
```

```python
X=np.array(x_train).reshape(-1, 1)
Y=np.array(y_train).reshape(-1, 1)

print(X.shape, Y.shape)
print(X[:5], Y[:5])
```

```python
lr=0.01
epooches=10000
loss_history=[]
```

```python
for i in range(epooches):
    Y_pred=X*W + B
    loss=np.mean((Y-Y_pred)**2)
    loss_history.append(loss)

    dj_dw=-2*np.mean(X*(Y-Y_pred))
    dj_db=-2*np.mean(Y-Y_pred)

    W-=lr*dj_dw
    B-=lr*dj_db

    if i % 100 == 0:
        print(f'Epoch {i}, Loss: {loss}, W: {W}, B: {B}')
    if i % 1000== 0:
        plt.scatter(X, Y, color='blue', label='Data Points')
        plt.plot(X, Y_pred, color='red', label='Prediction Line')
        plt.title(f'Epoch {i}')
        plt.xlabel('X')
        plt.ylabel('Y')
        plt.legend()
        plt.show()
   
    if i>100 and loss_history[-10]==loss :
        print(f'Loss did not improve in the last 10 epochs, stopping at epoch {i}')
        print(f'Final W: {W}, B: {B}, Loss: {loss}')
        break
        
```
