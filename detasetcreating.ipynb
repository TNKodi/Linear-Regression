```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import pandas as pd
```

### Create the random samples

```python
np.random.seed(42)
n_samples = 1000
X = np.round(np.random.rand(n_samples,) * 10, 4)

print(X[:10])  # Display first 10 samples
```

```python
W_true=3
b_true=4

print(f'Y={W_true} * X + {b_true}')
```

####   Add Gaussian noise

```python
noise = np.random.randn(n_samples) * 2
print(noise[:5])
```

Plot the Line and Noise data 

```python
Y= W_true * X + b_true + noise

plt.scatter(X[:100], Y[:100], label= 'Noisy Data', color='blue')
plt.plot(X, W_true * X + b_true, label='True Line', color='red')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Regression Data with Noise')
plt.legend()
plt.show()
```

```python
df=pd.DataFrame({
    'X': X,
    'Y': Y
})

df.to_csv('linear_regression_data.csv', index=False)

print("Data saved to 'linear_regression_data.csv'")
```
